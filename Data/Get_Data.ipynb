```python
import requests

api_key = ''
ETF_codes = []
ETF_names = []
L = [""]
i = 0
while len(L) > 0:
    print(i, end='\r')
    i += 1
    url = f'https://api.polygon.io/v3/reference/tickers?ticker.gt={L[-1]}&type=ETF&active=true&limit=1000&apiKey={api_key}'
    response = requests.get(url).json()
    L = []
    L2 = []
    if response["status"] != 'ERROR':
        L = [item["ticker"] for item in response["results"]]
        L2 = [item["name"] for item in response["results"]]
    ETF_codes.extend(L)
    ETF_names.extend(L2)

ETF_codes
```

```python
import yfinance as yf

etf_info = []
for k, ticker in enumerate(ETF_codes):
    print(ticker, end="\r")
    desc = yf.Ticker(ticker).info.get('longBusinessSummary', 'No description available')
    etf_info.append((ticker, ETF_names[k], desc))
print(etf_info)
```

```python
import pandas as pd
etf_df = pd.DataFrame(etf_info, columns=["code", "name", "description"])
etf_df.to_csv("ETF_info.csv")
```

```python
# List of stock tickers you want to fetch data for
tickers = ETF_codes

# Download historical data for these stocks
data = yf.download(tickers, start='2004-11-06', end='2024-11-07')

# Display the retrieved data
print(data)
```

```python
data.to_csv("ETF_values_data.csv")
```

```python
from yahooquery import Ticker

symbols = 'SPY IWDA.AS'
symbols = ETF_codes[:100]

tickers = Ticker(symbols)

tickers.fund_holding_info
```

```python
import pandas as pd
etf_data = []
for etf, info in tickers.fund_holding_info.items():
    # Flatten the info if it's nested (optional, depending on structure)
    if type(info) == str:
        continue
    flat_info = {f"{k}_{sub_k}" if isinstance(v, dict) else k: sub_v
                 for k, v in info.items()
                 for sub_k, sub_v in (v.items() if isinstance(v, dict) else [(None, v)])}
    flat_info['ETF'] = etf  # Add ETF code as a column
    etf_data.append(flat_info)

# Convert list of dictionaries to DataFrame
df = pd.DataFrame(etf_data).set_index('ETF')
df
```
