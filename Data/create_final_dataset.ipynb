```python
def extract_etf_data(etf_dict):
    # Extract top-level fields
    etf_info = {
        'maxAge': etf_dict.get('maxAge'),
        'cashPosition': etf_dict.get('cashPosition'),
        'stockPosition': etf_dict.get('stockPosition'),
        'bondPosition': etf_dict.get('bondPosition'),
        'otherPosition': etf_dict.get('otherPosition'),
        'preferredPosition': etf_dict.get('preferredPosition'),
        'convertiblePosition': etf_dict.get('convertiblePosition'),
    }

    # Extract equity holdings
    equity_holdings = etf_dict.get('equityHoldings', {})
    etf_info.update({
        'priceToEarnings': equity_holdings.get('priceToEarnings'),
        'priceToBook': equity_holdings.get('priceToBook'),
        'priceToSales': equity_holdings.get('priceToSales'),
        'priceToCashflow': equity_holdings.get('priceToCashflow'),
    })

    # Extract sector weightings
    sector_weightings = etf_dict.get('sectorWeightings', [])
    for sector in sector_weightings:
        for sector_name, weight in sector.items():
            etf_info[f'sector_{sector_name}'] = weight

    # Extract holdings (symbols, names, percentages as lists)
    holdings = etf_dict.get('holdings', [])
    etf_info['holdings_symbols'] = [h.get('symbol') for h in holdings]
    etf_info['holdings_names'] = [h.get('holdingName') for h in holdings]
    etf_info['holdings_percents'] = [h.get('holdingPercent') for h in holdings]

    return etf_info

df_extracted = pd.DataFrame(df_reduced['Holdings'].apply(extract_etf_data).tolist())
df_exposure = pd.concat([df_reduced, df_extracted], axis=1)
df_exposure
```

```python
from yahooquery import Ticker
def refine_stock_data(ticker):
    try:

        stock_data = Ticker(str(ticker)).history(period='5y')  
        return stock_data  
    except Exception as e:
        print(f"Error fetching data for ticker {ticker}: {e}")
        return None


df_exposure['Stock_Data'] = df_exposure['Ticker_Info'].apply(refine_stock_data)

df_exposure.to_pickle('df_exposure_with_refined_stock_data.pkl')

df_exposure_without_stock_data = df_exposure.drop(columns=['Stock_Data'])
df_exposure_without_stock_data.to_csv('df_exposure_metadata.csv', index=False)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import yfinance as yf

def compute_etf_metrics(stock_data, market_daily_return):
    if stock_data is None or stock_data.empty:
        return {
            'Var': None,
            'drawdown': None,
            'avg_sharpe_ratio': None,
            'beta': None,
            'avg_monthly_return': None,
            'volatility': None,
            'RSI': None,
        }
    
    # Reset index if MultiIndex and ensure 'date' column is the index
    if isinstance(stock_data.index, pd.MultiIndex):
        stock_data = stock_data.reset_index()
    if 'date' in stock_data.columns:
        stock_data['date'] = pd.to_datetime(stock_data['date'])
        stock_data.set_index('date', inplace=True)

    # Calculate daily returns
    stock_data['daily_return'] = stock_data['adjclose'].pct_change()

    # Value at Risk (VaR)
    confidence_level = 0.95
    stock_data['Var'] = stock_data['daily_return'].rolling(window=252).quantile(1 - confidence_level)

    # Drawdown
    stock_data['cum_return'] = (1 + stock_data['daily_return']).cumprod()
    stock_data['rolling_max'] = stock_data['cum_return'].cummax()
    stock_data['drawdown'] = stock_data['cum_return'] / stock_data['rolling_max'] - 1

    # Sharpe Ratio
    risk_free_rate = 0.01  # Example risk-free rate
    stock_data['sharpe_ratio'] = (stock_data['daily_return'].mean() - risk_free_rate) / stock_data['daily_return'].std()
    avg_sharpe_ratio = stock_data['sharpe_ratio'].mean()

    # CAPM Beta Calculation
    aligned_market_return = market_daily_return.reindex(stock_data.index).fillna(0)
    if aligned_market_return.shape[0] < 30 or stock_data['daily_return'].shape[0] < 30:
        beta = None  # Not enough data for regression
    else:
        # Drop NaN values for regression
        valid_data = stock_data[['daily_return']].join(aligned_market_return, how='inner', rsuffix='_market').dropna()
        if valid_data.shape[0] < 30:  # Ensure sufficient data
            beta = None
        else:
            X = valid_data['daily_return_market'].values.reshape(-1, 1)  # Market returns as independent variable
            Y = valid_data['daily_return'].values  # ETF returns as dependent variable
            reg = LinearRegression().fit(X, Y)
            beta = reg.coef_[0]  # Slope of the regression line

    # Average Monthly Return (rolling 30 days)
    stock_data['rolling_30d_return'] = stock_data['adjclose'].pct_change(periods=30)
    avg_monthly_return = stock_data['rolling_30d_return'].mean()

    # Volatility
    volatility = stock_data['daily_return'].rolling(window=252).std().mean()

    # RSI
    delta = stock_data['adjclose'].diff()
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = pd.Series(gain).rolling(window=14).mean()
    avg_loss = pd.Series(loss).rolling(window=14).mean()
    rs = avg_gain / avg_loss
    RSI = 100 - (100 / (1 + rs))
    RSI_last = RSI.iloc[-1] if not RSI.empty else None

    return {
        'Var': stock_data['Var'].iloc[-1] if not stock_data['Var'].empty else None,
        'drawdown': stock_data['drawdown'].min() if not stock_data['drawdown'].empty else None,
        'avg_sharpe_ratio': avg_sharpe_ratio,
        'beta': beta,
        'avg_monthly_return': avg_monthly_return,
        'volatility': volatility,
        'RSI': RSI_last,
    }

# Fetch market data (e.g., S&P 500)
market_ticker = '^GSPC'  # Example: S&P 500 Index
market_data = yf.download(market_ticker, start='2015-01-01', end='2024-12-31')

# Compute market daily returns
market_data['daily_return'] = market_data['Adj Close'].pct_change()
market_daily_return = market_data['daily_return']

# Iterate through df_exposure and calculate metrics for each ETF
metrics_list = []
for i in range(len(df_exposure)):
    stock_data = df_exposure['Stock_Data'][i]
    metrics = compute_etf_metrics(stock_data, market_daily_return)
    metrics_list.append(metrics)

# Create a DataFrame from the metrics and join with df_exposure
metrics_df = pd.DataFrame(metrics_list)
df_exposure = pd.concat([df_exposure, metrics_df], axis=1)


```
