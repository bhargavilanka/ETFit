```python
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('justetf.csv')

def fetch_ticker_info(isin):
    try:
        ticker = yf.Ticker(isin)
        return ticker 
    except Exception as e:
        print(f"Error fetching data for ISIN {isin}: {e}")
        return None

df['Ticker_Info'] = df['isin'].apply(fetch_ticker_info)
```

```python
df.to_csv('justetf_with_ticker_info.csv', index=False)
```

```python
import pandas as pd
from yahooquery import Ticker

df = pd.read_csv('justetf_with_ticker_info.csv')


def get_stock_data(ticker):
    return Ticker(str(ticker)).history(period='2y')

df['Stock_Data'] = df['Ticker_Info'].apply(get_stock_data)

df.to_csv('justetf_with_ticker_info_and_stock_data.csv', index=False)
```
